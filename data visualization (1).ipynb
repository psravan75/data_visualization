```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Set random seed for reproducibility
np.random.seed(30)

# Generate random data
data = {
    'X': np.random.uniform(-10, 10, 300),
    'Y': np.random.uniform(-10, 10, 300),
    'Z': np.random.uniform(-10, 10, 300)
}

# Create DataFrame
df = pd.DataFrame(data)

# Create figure and 3D subplot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Create 3D scatter plot
ax.scatter(df['X'], df['Y'], df['Z'])

# Add title
ax.set_title('3D Scatter Plot')

# Show the plot
plt.show()
```

```python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
import numpy as np

# Prepare the output for Jupyter Notebook
output_notebook()

# Generate random data
x = np.random.rand(100)  # Random x coordinates
y = np.random.rand(100)  # Random y coordinates
sizes = np.random.randint(10, 20, 100)  # Random sizes for each point
colors = np.random.randint(50, 150, 100)  # Random colors

# Normalize colors to be in the range [0, 255] and convert to uint8
colors = (colors - colors.min()) / (colors.max() - colors.min()) * 255
colors = colors.astype(np.uint8)  # Convert colors to uint8

# Create a figure
p = figure(title="Random Scatter Plot")

# Create scatter plot
p.scatter(x, y, size=sizes, color=colors, fill_alpha=0.6, line_color=None)

# Show the plot
show(p)
```

```python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook

# Prepare the output for Jupyter Notebook
output_notebook()

# Data for the bar chart
fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
counts = [20, 25, 30, 35]

# Create a figure with x_range set to the fruit names
p = figure(x_range=fruits, title="Fruit Counts", toolbar_location=None, tools="")

# Create vertical bars for the fruit counts
p.vbar(x=fruits, top=counts, width=0.4)

# Show the plot
show(p)
```

```python
import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_notebook

# Prepare the output for Jupyter Notebook
output_notebook()

# Generate random data
data_hist = np.random.randn(1000)

# Calculate histogram
hist, edges = np.histogram(data_hist, bins=30)

# Create a figure for the histogram
p = figure(title="Histogram", x_axis_label="Value", y_axis_label="Frequency")

# Create bars for the histogram using quad glyph
p.quad(top=hist, bottom=0, left=edges[:-1], right=edges[1:], 
        fill_color="blue", line_color="white", alpha=0.6)

# Show the plot
show(p)
```

```python
import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_notebook

# Prepare the output for Jupyter Notebook
output_notebook()

# Generate random data for the heatmap
data_heatmap = np.random.rand(10, 10)

# Create a figure for the heatmap
p = figure(title="Heatmap", x_axis_label="X", y_axis_label="Y", 
           width=400, height=400)

# Create the heatmap using the image method
p.image(image=[data_heatmap], x=0, y=0, dw=10, dh=10, palette="Viridis256")

# Show the plot
show(p)
```
